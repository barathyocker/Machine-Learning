```python
import pandas as pd
import numpy as py
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent= dataset[[ 'age', 'bmi', 'children', 'sex_male', 'smoker_yes' ]]
```

```python
dependent=dataset[['charges' ]]
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
independent= sc.fit_transform(independent)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel': ['rbf', 'poly','sigmoid','linear'],
             'C': [3000,5000,6000],'gamma':['auto','scale']}

grid =GridSearchCV(SVR(), param_grid, refit=True, verbose=3 ,n_jobs=-1)

```

```python
#fitting model for grid
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_
print("the R_score value for the best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```
