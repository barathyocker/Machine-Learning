```python
import pandas as pd
import numpy as py
import pickle
```

```python
dataset = pd.read_csv('CVD.csv')
```

```python
dataset
```

```python
dataset.dtypes
```

```python
independent=dataset[['gender', 'chestpain', 'restingBP', 'restingrelectro', 'oldpeak']]
```

```python
dependent=dataset[['target' ]]
```

```python
independent
```

```python
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
independent= sc.fit_transform(independent)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

param_grid = {'criterion': ['gini', 'entropy','log_loss'],
             'max_features':['sqrt','log2'], 'n_estimators' : [10,50,100] }

grid =GridSearchCV(RandomForestClassifier(), param_grid, refit=True, verbose=3 ,n_jobs=-1)
```

```python
grid.fit(independent,dependent)
```

```python
re=grid.cv_results_
print("the R_score value for the best parameter {}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```

```python
table
```

```python
gender= int(input('male=1 or female = 0'))
chestpain=int(input('Value 0: typical angina or Value 1: atypical angina or Value 2: non-anginal pain or Value 3: asymptomatic'))
restingBP=int(input('94-200 (in mm HG)'))
restingrelectro=int(input("Value 0: normal or Value 1: having ST-T wave abnormality or Value 2: showing probable or definite left ventricularhypertrophy by Estes criteria"))
oldpeak=float(input("value 0-6.2"))
```

```python
scx=sc.transform
sc_x=scx([[gender,chestpain,restingBP,restingrelectro,oldpeak]])
```

```python
future_prediction=grid.predict(sc_x)
print("future_predicition={}".format(future_prediction))
```

```python
grid_predictions=grid.predict(independent)
from sklearn.metrics import classification_report
clf_report = classification_report(dependent, grid_predictions)
print(clf_report)
```

```python
import pickle # for saving the best model from grid search cv
filename="CVD_finalized_RF_model.sav"
pickle.dump(grid,open(filename,'wb'))
```
