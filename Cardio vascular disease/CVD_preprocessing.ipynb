```python
import pandas as pd
```

```python
dataset = pd.read_csv('CVD.csv')
```

```python
dataset= dataset.drop('patientid',axis=1)
```

```python
dataset
```

```python
#checking for null values
dataset.isnull().sum()
```

```python
table = pd.DataFrame(index=['mean','median','mode','Q1','Q2','Q3','Q4','IQR','lesser out'
                           ,'min','greater out','max',"kurtosis",'skew','var','std'],columns=dataset.columns)
```

```python
table
```

```python
for clnm in dataset.columns:
    table[clnm]['mean']= dataset[clnm].mean()
    table[clnm]['median']= dataset[clnm].median()
    table[clnm]['mode']= dataset[clnm].mode()[0]
    table[clnm]['Q1']= dataset.describe()[clnm]['25%']
    table[clnm]['Q2']= dataset.describe()[clnm]['50%']
    table[clnm]['Q3']= dataset.describe()[clnm]['75%']
    table[clnm]['Q4']= dataset.describe()[clnm]['max']
    table[clnm]['min']= min(dataset[clnm])
    table[clnm]['max']= max(dataset[clnm])
    table[clnm]['IQR']=table[clnm]['Q3'] - table[clnm]['Q1']
    table[clnm]['lesser out']= dataset.describe()[clnm]['25%'] - (1.5* table[clnm]['IQR'])
    table[clnm]['greater out']= dataset.describe()[clnm]['75%'] + (1.5* table[clnm]['IQR'])
    table[clnm]["kurtosis"]=dataset[clnm].kurtosis()
    table[clnm]["skew"]=dataset[clnm].skew()
    table[clnm]["var"]=dataset[clnm].var()
    table[clnm]["std"]=dataset[clnm].std()
```

```python
table
```

```python
lesser_diff=[]
greater_diff=[]
```

```python
for clnm in dataset.columns:
    if table[clnm]['lesser out'] > table[clnm]['min']:
        lesser_diff.append(clnm)
    if table[clnm]['max'] > table[clnm]['greater out']:
        greater_diff.append(clnm)
```

```python
#outlier found in gender column. it can be avoided
lesser_diff
```

```python
#frequency
def freq (clnm,dataset):
    
    freqtab= pd.DataFrame(columns=['unique_values','frequency','rel_frequency','cum_frequency'])
   
    freqtab ['unique_values'] = dataset[clnm].value_counts().index 
    freqtab ['frequency'] = dataset[clnm].value_counts().values
    freqtab ['rel_frequency'] = freqtab ['frequency']/dataset[clnm].count()
    freqtab ['cum_frequency'] = freqtab ['rel_frequency'].cumsum()
    
    return freqtab
```

```python
freq('target',dataset)
```
